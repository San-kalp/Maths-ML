```python
def gradient_descent(dfdx, dfdy, x , y , learning_rate = 0.1 , n = 100):
    for iterations in range(n):
        x = x - learning_rate * dfdx(x,y)
        y = y - learning_rate * dfdy(x,y)
    return x , y
```
