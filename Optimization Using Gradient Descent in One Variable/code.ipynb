```python
import numpy as np
import matplotlib.pyplot as plt 
from tools import  plot_f, gradient_descent_one_variable, f_example_2, dfdx_example_2
```

```python
def f_of_x(x):
    return np.exp(x) - np.log(x)

def dfdx(x):
    return np.exp(x) - 1/x
```

```python
plot_f([0.001, 2.5], [-0.3, 13], f_of_x, 0.0)
```

```python
def gradient_descent(dfdx, x, learning_rate = 0.1, num_iterations = 100):
    for iteration in range(num_iterations):
        x = x - learning_rate * dfdx(x)
    return x
```

```python
gd = gradient_descent(dfdx , 1.6 , 0.1 , 25)

print(gd)
```
